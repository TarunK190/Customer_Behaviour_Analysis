```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns=df.columns.str.lower()  
df.columns=df.columns.str.replace(' ','_') 
df.columns=df.columns.str.replace('purchase_amount_(usd)','purchase_amount')    
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.head()
df.columns
```

```python
#create column age_group
labels=['Young_adult','Adult','Middle_aged','Senior','high']
df['age_group']=pd.qcut(df['age'],q=5,labels=labels)
```

```python
df[['age','age_group']].head(50)
```

```python
df['frequency_of_purchases'].unique()
```

```python
#create frequency of purchase frequency days

frequency_mapping={
    'Fortnightly':14,
    'Weekly':7,
    'Annually':365,
    'Quarterly':90,
    'Bi-Weekly':90,
    'Monthly':30,
    'Every 3 Months':90

}

df['purchase_frequency_days']=df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(50)
```
